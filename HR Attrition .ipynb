```python
# remove unwanted columns & splitting the data in x and y

x=a.drop(labels=["Attrition","EmployeeCount","EmployeeNumber","StandardHours","Over18"],axis=1)
y=a[["Attrition"]]
```

```python
cat=[]
con=[]
for i in x.columns:
    if(x[i].dtypes=="object"):
        cat.append(i)
    else:
        con.append(i)
```

```python
cat
```

```python
#converting all the string values in integer 

from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
for i in x[cat]:
    x[i]=lb.fit_transform(x[i])
```

```python
#Standerdzing the dataset at same scale

from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x=pd.DataFrame(ss.fit_transform(x[con]),columns=con)
```

```python
x
```

# Model making

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier(criterion='entropy',random_state=1)
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
z=[rfc,dtc,lr]
```

```python
    for i in z:
        model=i.fit(xtrain,ytrain)
        pred_ts=model.predict(xtest)
        pred_tr=model.predict(xtrain)
        from sklearn.metrics import accuracy_score
        ts_acc = round(accuracy_score(ytest,pred_ts),3)
        tr_acc = round(accuracy_score(ytrain,pred_tr),3)
        print(i)
        print("training accuracy: ",tr_acc,"\ntesting_accuracy: ",ts_acc)
        if(tr_acc>ts_acc):
            print("Model overfits")
        else:
            print("No overfitting")
        print("-------------------")

```

# Tuning the model to find best accuracy

```python
def tuning(m):
    model=m.fit(xtrain,ytrain)
    pred_ts=model.predict(xtest)
    pred_tr=model.predict(xtrain)
    from sklearn.metrics import accuracy_score
    ts_acc = round(accuracy_score(ytest,pred_ts),3)
    tr_acc = round(accuracy_score(ytrain,pred_tr),3)
    print(i)
    print("training accuracy: ",tr_acc,"\ntesting_accuracy: ",ts_acc)
    if(tr_acc>ts_acc):
        print("Model overfits")
    else:
        print("No overfitting")
   
```

# Tuning for Decision tree

```python
#best test accuracy 0.84 for max_depth 2
```

```python
for i in range(2,18,1):
    dtc=DecisionTreeClassifier(criterion='entropy',random_state=1,max_depth=i)
    print("---------------------------")
    print("Results for max depth: %d"%i)
    tuning(dtc)
```

# Tuning for Random forest classification

```python
#best test accuracy 0.853 for max_depth 7
```

```python
for i in range(2,18,1):
    rfc=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0,max_depth=i)
    print("---------------------------")
    print("Results for max depth: %d"%i)
    tuning(rfc)
```
